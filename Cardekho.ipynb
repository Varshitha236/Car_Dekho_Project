```python
import pandas as pd
```

```python
df=pd.read_csv('car_Dekho_DA.csv')
df.head()
```

```python
df.info()
```

```python
df=pd.DataFrame(df)
```

```python
# 1)From which manufacturing year to which manufacturing year vehicles are present in this data?
manufacturing_year_range = df['Year'].min(), df['Year'].max()
print("Manufacturing year range:", manufacturing_year_range)
```

```python
# 2)What is the lowest price to which a vehicle is sold?
lowest_price = df['Selling_Price'].min()
print("Lowest price sold:", lowest_price, "lakhs")
```

```python
# 3)What is the highest price to which a vehicle is sold?
highest_price = df['Selling_Price'].max()
print("Highest price sold:", highest_price, "lakhs")
```

```python
# 4)How many records are there in this data?
total_records = df.shape[0]
print("Total records:", total_records)
```

```python
# 5)Are there any missing records in this data?
missing_records = df.isnull().sum().sum()
print("Missing records:", missing_records)
```

```python
# 6)How many different vehicles are present in this data?
unique_vehicles = df['Car_Name'].nunique()
print("Different vehicles:", unique_vehicles)
```

```python
#7)Which is the most sold vehicle in this data ?
most_sold_car=df["Car_Name"].mode()
print(f"the most sold car is", most_sold_car)
```

```python
# 8)Does the database include any CNG vehicle? If yes, how many of them are there?
cng_vehicles_count = df[df['Fuel_Type'] == 'CNG'].shape[0]
print("CNG vehicles count:", cng_vehicles_count)
```

```python
# 9)How many vehicles here are for sale from Individuals directly?
individual_sales_count = df[df['Seller_Type'] == 'Individual'].shape[0]
print("Vehicles for sale by Individuals:", individual_sales_count)
```

```python
# 10)Does this database contain auto transmission vehicles? If yes, how many of them are there?
df['Transmission'].unique()
df['Transmission'].value_counts()
```

```python
# 11)How many single person owned vehicles are there in this database?
df['Owner'].value_counts()
```

```python
# 12)Which is the most and least cost depreciated vehicle in data?
df['Depreciation'] = df['Present_Price'] - df['Selling_Price']
min_depreciation = df['Depreciation'].min()
df[df['Depreciation'] == min_depreciation]
```

```python
# 12)Highest depreciated vehicle
max_depreciation = df['Depreciation'].max()
df[df['Depreciation'] == max_depreciation]
```

```python
# 13)Which brands of vehicles are less affected by cost depreciation?
min_depreciation = df['Depreciation'].min()
df[df['Depreciation'] == min_depreciation]
```

```python
# 14) Are there any factors which you feel affect the cost depreciation?
df['Depreciation']
```

```python
# 15) In general, selling price is affected by age of vehicle and distance driven by vehicle, is it observable from data?
import matplotlib.pyplot as plt
plt.scatter(df['Year'],df['Selling_Price'])
plt.xlabel('Year')
plt.ylabel('Selling Price')
plt.show()
```

```python
# 16)Can we get idea about newest vehicles i.e. after 2014 manufactured?
newest_vehicles = df[df['Year'] > 2014]
print("Newest vehicles after 2014:\n", newest_vehicles)
```

```python
#Fuel Type analysis
import matplotlib.pyplot as plt
plt.pie(df.Fuel_Type.value_counts(),labels = ["Petrol", "Diesel" , "CNG"],shadow = True)
plt.show()
```

```python
# Plot Selling Price Distribution
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(10, 6))
sns.histplot(df['Selling_Price'], bins=30, kde=True)
plt.title('Selling Price Distribution')
plt.xlabel('Selling Price')
plt.ylabel('Frequency')
plt.show()
```

```python
# Plot Kms Driven vs. Selling Price
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Kms_Driven', y='Selling_Price', data=df)
plt.title('Kms Driven vs. Selling Price')
plt.xlabel('Kms Driven')
plt.ylabel('Selling Price')
plt.show()
```
